```python
from functools import lru_cache
total_calls = 0
@lru_cache(maxsize=None)
def fib(n: int) -> int:
    global total_calls
    total_calls += 1
    print(f"fib(n={n})")
    return n if n <= 1 else fib(n-1) + fib(n-2)
```
